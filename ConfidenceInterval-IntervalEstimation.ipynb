```python
import pandas as pd
import numpy as np
import scipy.stats as st
import math

pd.set_option('display.max_columns', 30)
df=pd.read_csv('S1_Dataset.csv')
df
```

```python
#Our goal is to analyze the bedtime of napping and non-napping toddlers.
#night_bedtime=df['night bedtime']
#daytime_nap=df.loc[df.napping==1]
bedtime_nap=df['night bedtime'].loc[df['napping']==1]
bedtime_no_nap=df['night bedtime'].loc[df['napping']==0]
print(len(bedtime_nap))
print(len(bedtime_no_nap))



```

```python


CIbedtimeForNaptaken=st.t.interval(alpha=0.95, df=len(bedtime_nap)-1, loc=np.mean(bedtime_nap), scale=st.sem(bedtime_nap)) 

CIbedtimeForNapNottaken=st.t.interval(alpha=0.95, df=len(bedtime_no_nap)-1, loc=np.mean(bedtime_no_nap), scale=st.sem(bedtime_no_nap)) 
#st.norm.interval(alpha=0.95, loc=np.mean(bedtime_nap), scale=st.sem(bedtime_nap))
print(CIbedtimeForNaptaken)
print(CIbedtimeForNapNottaken)



    
```

# Interpretation: 

From the above results, we conclude that we are 95% confident that the average bedtime for napping toddlers is between the time 19.98 – 20.63 (pm) while for non-napping toddlers it is between 18.96 – 20.22 (pm). These results are as per our expectation that if you take a nap during the day you will sleep late at night.

```python


def interval_estimate(n,data):
    try:
        if (n>=30):
            CI=st.norm.interval(alpha=0.95, loc=np.mean(data), scale=st.sem(data))
        else:
            CI=st.t.interval(alpha=0.95, df=len(data)-1, loc=np.mean(data), scale=st.sem(data))
            #print(CI)
        return CI
    except:
        print("check the type of estimate an dits proper calculation")
            
```

```python
n=50
data = np.random.randint(10, 30, n) # in the range of 10 to 30 take 50 count
CI=interval_estimate(n,data)  
print(f"The confidence interval is {CI}")
```
