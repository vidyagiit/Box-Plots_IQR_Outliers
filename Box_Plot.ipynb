```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

salaries = [15000, 15000, 20000, 20000, 20000, 25000, 25000, 30000, 35000, 200000]
salaries_df=pd.DataFrame(salaries,columns=['salaries'])
plt.subplots(figsize=(30,20))
#sns.boxplot(data=salaries_df,x='salaries')
sns.distplot(salaries_df)
plt.xticks(np.arange(min(salaries), max(salaries)+1, 10000))
#ax.xaxis.set_ticks(np.arange(start, end, stepsize))
plt.grid()
#plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker

salaries = [15000, 15000, 20000, 20000, 20000, 25000, 25000, 30000, 35000, 200000]
salaries_df=pd.DataFrame(salaries,columns=['salaries'])
fig, ax = plt.subplots()
sns.boxplot(data=salaries_df,x='salaries')
loc = plticker.MultipleLocator(base=1.0) # this locator puts ticks at regular intervals
ax.xaxis.set_major_locator(loc)
plt.show()
```
